## VALUE CHECK

Value check disini akan dibagi dengan melakukan value check kolom numerik dan kategorikal. meliputi value.count() dan nunique()

### Numerik


```python
# Age
print("--- Kolom: Age ---")
print(df['Age'].value_counts(dropna=False))
print(f"Jumlah unique: {df['Age'].nunique(dropna=False)}\n")
```

```python
# Height
print("--- Kolom: Height ---")
print(df['Height'].value_counts(dropna=False))
print(f"Jumlah unique: {df['Height'].nunique(dropna=False)}\n")
```

```python
# Weight
print("--- Kolom: Weight ---")
print(df['Weight'].value_counts(dropna=False))
print(f"Jumlah unique: {df['Weight'].nunique(dropna=False)}\n")

```

```python
# FCVC
print("--- Kolom: FCVC ---")
print(df['FCVC'].value_counts(dropna=False))
print(f"Jumlah unique: {df['FCVC'].nunique(dropna=False)}\n")

```

```python
# NCP
print("--- Kolom: NCP ---")
print(df['NCP'].value_counts(dropna=False))
print(f"Jumlah unique: {df['NCP'].nunique(dropna=False)}\n")

```

```python
# CH2O
print("--- Kolom: CH2O ---")
print(df['CH2O'].value_counts(dropna=False))
print(f"Jumlah unique: {df['CH2O'].nunique(dropna=False)}\n")
```

```python
# Ganti semua tanda '?' menjadi NaN
df.replace('?', np.nan, inplace=True)

# Konversi kolom numerik ke float
numerik_cols = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF']
for col in numerik_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # errors='coerce' juga akan ubah nilai invalid jadi NaN
```

##

```python
import seaborn as sns
import matplotlib.pyplot as plt

for col in numerik_cols:
    plt.figure(figsize=(12, 4))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Distribusi {col}")

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot {col}")

    plt.tight_layout()
    plt.show()
```

## kolom numerik ke target


```python
for col in numerik_cols:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='NObeyesdad', y=col, data=df)
    plt.title(f"{col} vs Kategori Obesitas")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
