## **VALUE CHECK**

Value check disini akan dibagi dengan melakukan value check kolom numerik dan kategorikal. meliputi value.count() dan nunique()

Value check akan terdiri dari 3 bagian fitur :
- Numerik
- Kategorikal Biner
- Kategorikal non Biner

```python
# NCP
print("--- Kolom: NCP ---")
print(df['NCP'].value_counts(dropna=False))
print(f"Jumlah unique: {df['NCP'].nunique(dropna=False)}\n")

```

```python
# Ganti semua tanda '?' menjadi NaN
df.replace('?', np.nan, inplace=True)

# Konversi kolom numerik ke float
numerik_cols = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']
for col in numerik_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # errors='coerce' juga akan ubah nilai invalid jadi NaN
```

## Visualisasi kolom numerik

```python
import seaborn as sns
import matplotlib.pyplot as plt

for col in numerik_cols:
    plt.figure(figsize=(12, 4))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Distribusi {col}")

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot {col}")

    plt.tight_layout()
    plt.show()
```

## kolom numerik ke target


```python
for col in numerik_cols:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='NObeyesdad', y=col, data=df)
    plt.title(f"{col} vs Kategori Obesitas")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# FAVC
print("--- Kolom: FAVC ---")
print(df['FAVC'].value_counts(dropna=False))
print(f"Jumlah unique: {df['FAVC'].nunique(dropna=False)}\n")

```

```python
# SMOKE
print("--- Kolom: SMOKE ---")
print(df['SMOKE'].value_counts(dropna=False))
print(f"Jumlah unique: {df['SMOKE'].nunique(dropna=False)}\n")

```

```python
# SCC
print("--- Kolom: SCC ---")
print(df['SCC'].value_counts(dropna=False))
print(f"Jumlah unique: {df['SCC'].nunique(dropna=False)}\n")
```

```python
# Fitur kategorikal biner
biner_cols = ['Gender', 'fam_history', 'FAVC', 'SMOKE', 'SCC']


for col in biner_cols:
    plt.figure(figsize=(12, 5))

    # Distribusi umum
    plt.subplot(1, 2, 1)
    sns.countplot(x=col, data=df, order=df[col].value_counts().index)
    plt.title(f"Distribusi {col}")
    plt.xticks(rotation=45)

    # Distribusi per kelas target
    plt.subplot(1, 2, 2)
    sns.countplot(x=col, hue='NObeyesdad', data=df, order=df[col].value_counts().index)
    plt.title(f"{col} vs Kategori Obesitas")
    plt.xticks(rotation=45)
    plt.legend(title='NObeyesdad')

    plt.tight_layout()
    plt.show()
```

## Keseimbangan Data Label Nobayesdad


```python
# Cek distribusi label
print(df['NObeyesdad'].value_counts())

# (Opsional) Tambahkan persentase untuk melihat proporsi
print("\nDistribusi dalam Persentase:")
print(df['NObeyesdad'].value_counts(normalize=True) * 100)


plt.figure(figsize=(8, 5))
sns.countplot(x='NObeyesdad', data=df, order=df['NObeyesdad'].value_counts().index)
plt.title("Distribusi Label 'NObeyesdad'")
plt.xticks(rotation=45)
plt.ylabel("Jumlah")
plt.xlabel("Kategori Obesitas")
plt.tight_layout()
plt.show()
```

dari visualisasi Distribusi label NObeyesdad dalam grafik terlihat cukup seimbang, meskipun ada sedikit perbedaan antara kelas terbanyak (Obesity_Type_I) dan paling sedikit (Insufficient_Weight). Ketidakseimbangan ini tidak terlalu ekstrem
